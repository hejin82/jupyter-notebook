```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(0, 10, num=11)
print(type(x1), x1)
x2 = np.arange(0, 11, 1)
print(type(x2), x2)
x3 = np.arange(11)
print(type(x3), x3)
```

```python
xpt = np.linspace(0, 10, 500)
ypt1 = np.sin(xpt)
ypt2 = np.cos(xpt)
plt.scatter(xpt, ypt1, label="sin")
plt.scatter(xpt, ypt2, label="cos")
plt.legend()
plt.show()
```

```python
left = -2 * np.pi
right = 2 * np.pi
x = np.linspace(left, right, 100)
f1 = 2 * np.sin(x)
f2 = 2 * np.sin(2 * x)
f3 = 0.5 * np.sin(x)
plt.plot(x, f1)
plt.plot(x, f2)
plt.plot(x, f3)
plt.show()
```

```python
xpt = np.linspace(0, 5, 500)
ypt = 1 - 0.5 * np.abs(xpt - 2)
lwidths = (1 + xpt) ** 2
plt.scatter(xpt, ypt, s=lwidths, color=(0, 1, 0))
plt.show()
```

```python
left = -np.pi
right = 2 * np.pi
x = np.linspace(left, right, 100)
y = np.sin(3 * x)
plt.plot(x, y)
plt.fill_between(x, 0, y, color='green', alpha=0.2)
plt.show()
```

```python
left = -np.pi
right = np.pi
x = np.linspace(left, right, 100)
y = np.sin(3 * x)
plt.plot(x, y)
plt.fill_between(x, -1, y, color='green', alpha=0.2)
plt.show()
```

```python
x = np.arange(100)
y = x
t = x
plt.scatter(x, y, c=t, cmap='rainbow')
plt.show()
```
