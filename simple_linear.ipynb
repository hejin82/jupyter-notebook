```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])
```

```python
plt.scatter(x, y)
plt.axis([0, 6, 0, 6])
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
num = 0.0
d = 0.0
for x_i, y_i in zip(x, y):
    num += (x_i - x_mean) * (y_i - y_mean)
    d += (x_i - x_mean) ** 2
```

```python
a = num / d
b = y_mean - a * x_mean
```

```python
y_hat = a * x + b
plt.scatter(x, y)
plt.plot(x, y_hat)
plt.axis([0, 6, 0, 6])
```
