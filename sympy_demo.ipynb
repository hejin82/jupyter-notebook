```python
from sympy.plotting import plot
from sympy import Symbol
```

```python
x = Symbol('x')
plot((2 * x -5), (x, -5, 5), title='Sympy', xlabel='x', ylabel='2x-5')
```

```python
x = Symbol('x')
plot(2*x - 5, 3*x + 2)
```

```python
x = Symbol('x')
line = plot(2*x-5, 3*x+2, show=False)
line[1].line_color = 'red'
line.show()
```

```python
x = Symbol('x')
line = plot(2*x-5, 3*x+2, show=False, legend=True)
line[1].line_color = 'red'
line.show()
```
