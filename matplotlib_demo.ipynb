```python
import matplotlib.pyplot as plt
```

```python
x = [x for x in range(9)]
squares = [0,1,4,9,16,25,36,49,64]
plt.plot(x,squares)
plt.show()
```

```python
squares = [0,1,4,9,16,25,36,49,64]
plt.plot(squares)
plt.show()
```

```python
squares = [0,1,4,9,16,25,36,49,64]
plt.plot(squares)
plt.axis([0,8,0,70])
plt.show()
```

```python
squares = [0,1,4,9,16,25,36,49,64]
plt.plot(squares)
plt.axis([0,8,0,70])
plt.grid()
plt.show()
```

```python
squares = [0,1,4,9,16,25,36,49,64]
plt.plot(squares, linewidth=10)
plt.axis([0,8,0,70])
plt.grid()
plt.show()
```

```python
squares = [0,1,4,9,16,25,36,49,64]
plt.plot(squares, linewidth=10)
plt.axis([0,8,0,70])
plt.grid()
plt.title('Test Chart', fontsize=24)
plt.xlabel('Value', fontsize=16)
plt.ylabel('Square', fontsize=16)
plt.show()
```

```python
squares = [0,1,4,9,16,25,36,49,64]
plt.plot(squares, linewidth=10)
plt.axis([0,8,0,70])
plt.grid()
plt.tick_params(axis='both', labelsize=14, color='red')
plt.title('Test Chart', fontsize=24)
plt.xlabel('Value', fontsize=16)
plt.ylabel('Square', fontsize=16)
plt.show()
```
