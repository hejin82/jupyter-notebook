```python
a = sympy.Symbol('a')
b = sympy.Symbol('b')
eq1 = a + b -1
eq2 = 5 * a + b -2
ans = sympy.solve((eq1, eq2))
print(type(ans))
print('ans = {}'.format(ans))
print('a = {}'.format(ans[a]))
print('b = {}'.format(ans[b]))
```

```python
a = sympy.Symbol('a')
b = sympy.Symbol('b')
eq1 = a + b -1
eq2 = 5 * a + b -2
ans = sympy.solve((eq1, eq2))
print(type(ans))
print('ans = {}'.format(ans))
print('a = {}'.format(ans[a]))
print('b = {}'.format(ans[b]))
pt_x1 = 600
pt_y1 = ans[a] * pt_x1 + ans[b]
pt_x2 = 1000
pt_y2 = ans[a] * pt_x2 + ans[b]

x = np.linspace(0, 2500, 250)
y = ans[a] * x + ans[b]
plt.plot(x, y)
plt.plot(pt_x1, pt_y1, '-o')
plt.text(pt_x1+60, pt_y1-10, 'pt1')
plt.plot(pt_x2, pt_y2, '-o')
plt.text(pt_x1+60, pt_y1-10, 'pt2')
plt.xlabel('Customers')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
a = 0.03
b = -18
x = np.linspace(0, 2500, 250)
y = a * x + b
pt_x = 1500
pt_y = a * pt_x + b
print('f(1500) = {}'.format(pt_y))
plt.plot(x, y)
plt.plot(pt_x, pt_y, '-o')
plt.text(pt_x-150, pt_y+3, 'f(1500)')
plt.xlabel('Customers')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
x = sympy.Symbol('x')
y = sympy.Symbol('y')
eq1 = x + y -35
eq2 = 2 * x + 4 * y - 100
ans = sympy.solve((eq1, eq2))
print('ans = {}'.format(ans))

line1_x = np.linspace(0, 100, 100)
line1_y = [35 - y for y in line1_x]
line2_x = np.linspace(0, 100, 100)
line2_y = [25 - 0.5 * y for y in line2_x]

plt.plot(line1_x, line1_y)
plt.plot(line2_x, line2_y)

plt.plot(ans[x], ans[y], '-o')
plt.text(ans[x]-5, ans[y]+5, '(' + str(ans[x]) + ', ' + str(ans[y]) + ')')
plt.xlabel('Chicken')
plt.ylabel('Rabbit')
plt.grid()
plt.show()
```

```python
x = sympy.Symbol('x')
y = sympy.Symbol('y')
eq1 = x - y
eq2 = -x - y + 2
ans = sympy.solve((eq1, eq2))
print('ans = {}'.format(ans))
line1_x = np.linspace(-5, 5, 10)
line1_y = [y for y in line1_x]
line2_x = np.linspace(-5, 5, 10)
line2_y = [-y + 2 for y in line2_x]
plt.plot(line1_x, line1_y)
plt.plot(line2_x, line2_y)

plt.plot(ans[x], ans[y], '-o')
plt.text(ans[x]-0.5, ans[y]+0.3, '(' + str(ans[x]) + ', ' + str(ans[y]) + ')')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axis('equal')
plt.grid()
plt.show()
```

```python
x = sympy.Symbol('x')
y = sympy.Symbol('y')
eq1 = 0.5 * x - y - 0.5
eq2 = -2 * x - y + 7
ans = sympy.solve((eq1, eq2))
print('ans = {}'.format(ans))
line1_x = np.linspace(-5, 5, 10)
line1_y = [(0.5 * y - 0.5) for y in line1_x]
line2_x = np.linspace(-5, 5, 10)
line2_y = [(-2 * y + 7) for y in line2_x]
plt.plot(line1_x, line1_y)
plt.plot(line2_x, line2_y)

plt.plot(ans[x], ans[y], '-o')
plt.text(ans[x]-0.7, ans[y]+0.5, '(' + str(ans[x]) + ', ' + str(ans[y]) + ')')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axis('equal')
plt.grid()
plt.show()
```
